# Exercice 1 — Explorer et décrire un jeu de données

**Difficulté :** Débutant  
**Durée estimée :** 20 min  

---

## Contexte

On reçoit un export CSV provenant d'un réseau de capteurs IoT installés dans un entrepôt industriel. Le fichier `capteurs_temperature.csv` contient des relevés de température et d'humidité enregistrés automatiquement toutes les 15 minutes.

Avant toute analyse, la première étape consiste toujours à comprendre ce que contient le jeu de données : sa structure, ses dimensions, ses types de données, et ses éventuelles anomalies.

**Objectif :** Réaliser une exploration complète d'un CSV et en extraire une description synthétique.

---

## Données

Fichier : `../data/capteurs_temperature.csv`

---

## Tâche 1 — Charger le fichier avec pandas

Importer la bibliothèque pandas et charger le fichier CSV dans un DataFrame.

## Tâche 2 — Afficher les 10 premières lignes

Observer les premières lignes pour comprendre la structure des données.

## Tâche 3 — Combien de lignes et de colonnes ?

Afficher les dimensions du DataFrame (nombre de lignes et nombre de colonnes).

## Tâche 4 — Quels sont les types de chaque colonne ?

Afficher le type de données (dtype) de chaque colonne. Identifier si les types sont cohérents avec ce qu'on attendrait.

## Tâche 5 — Y a-t-il des valeurs manquantes ?

Compter le nombre de valeurs nulles par colonne. Si des valeurs manquent, indiquer dans quelle proportion.

## Tâche 6 — Quels capteurs sont présents dans le jeu de données ?

Lister les identifiants uniques de capteurs et indiquer combien de relevés chacun a produit.

## Tâche 7 — Quelle est la plage de dates couverte ?

Convertir la colonne `timestamp` en type datetime, puis afficher la date de début et la date de fin de la période couverte.

## Tâche 8 — Statistiques descriptives

Calculer les statistiques descriptives (min, max, moyenne, écart-type) pour les colonnes `temperature_c` et `humidite_pct`.

## Tâche 9 — Identifier le capteur avec la température moyenne la plus élevée

Calculer la température moyenne par capteur et identifier lequel a la valeur la plus élevée.

## Tâche 10 — Bonus : repérer un comportement anormal

Un capteur a-t-il des relevés de température anormalement élevés ou faibles ? Utiliser l'écart-type pour définir un seuil d'anomalie (par exemple : valeur > moyenne + 2 × écart-type).
