# Exercice 3 — De la donnée brute au rapport d'analyse

**Difficulté :** Avancé  
**Durée estimée :** 60 min  

---

## Contexte

Le responsable énergie du groupe demande une analyse complète des consommations énergétiques pour préparer un plan de réduction des coûts. Il dispose du fichier `consommation_energie.csv` qui couvre deux années de relevés quotidiens pour plusieurs bâtiments.

Cet exercice simule une mission d'analyse complète : de la donnée brute jusqu'au rapport structuré et aux recommandations opérationnelles. C'est le type de livrable attendu dans un contexte professionnel.

**Objectif :** Produire une analyse complète avec exploration, nettoyage, KPIs, visualisations et recommandations exportables.

---

## Données

Fichier : `../data/consommation_energie.csv`

Colonnes disponibles :
- `date` — jour du relevé
- `batiment` — site (Usine_A, Usine_B, Bureau_Central, Entrepot_Nord)
- `zone` — zone fonctionnelle (Production, Eclairage, Climatisation, Bureaux)
- `type_energie` — vecteur énergétique (electricite, gaz)
- `kwh` — consommation en kilowattheures
