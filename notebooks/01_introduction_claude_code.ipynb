Ce notebook pose les fondations conceptuelles avant toute manipulation technique. L'objectif est de comprendre précisément ce qu'est Claude Code, comment il fonctionne, et dans quels contextes professionnels il apporte une valeur réelle.

## 1. Qu'est-ce que Claude Code ?

Claude Code est un **agent de code** développé par Anthropic. Il s'exécute dans le terminal et interagit directement avec le système de fichiers, les outils de développement et l'environnement de travail.

Il est important de ne pas le confondre avec un chatbot classique. La différence est fondamentale en pratique.

### 1.1 Agent vs Chatbot — Tableau comparatif

| Dimension | Chatbot classique (Claude.ai) | Agent de code (Claude Code) |
|-----------|-------------------------------|-----------------------------|
| **Interface** | Navigateur web | Terminal (ligne de commande) |
| **Accès aux fichiers** | Non — fonctionne uniquement avec ce qu'on lui colle | Oui — lit et écrit directement dans les fichiers du projet |
| **Exécution de code** | Non — génère du code mais ne peut pas l'exécuter | Oui — exécute les scripts, vérifie les résultats, corrige les erreurs |
| **Mémoire du projet** | Aucune — chaque conversation repart de zéro | Contextuelle — explore la structure du projet pour comprendre l'existant |
| **Actions possibles** | Texte, code, analyse (statique) | Créer, modifier, supprimer des fichiers ; lancer des commandes |
| **Autonomie** | Répond à une question | Accomplit une tâche de bout en bout |
| **Vérification** | Impossible — l'utilisateur doit vérifier manuellement | Intégrée — l'agent teste, observe les erreurs, itère |
| **Usage typique** | Expliquer un concept, rédiger un email, reformuler | Analyser un CSV, construire un dashboard, automatiser un traitement |

> **Point clé** : un chatbot répond, un agent agit. Cette distinction change radicalement la façon de travailler.

### 1.2 Anatomie d'un agent

Un agent est un système qui peut :

1. **Percevoir** son environnement (lire des fichiers, observer des erreurs, inspecter une structure de dossiers)
2. **Raisonner** sur ce qu'il perçoit (comprendre un problème, planifier une solution)
3. **Agir** sur son environnement (créer un fichier, exécuter une commande, modifier du code)
4. **Observer** les conséquences de ses actions (lire la sortie d'erreur, vérifier que le fichier a bien été créé)
5. **Itérer** jusqu'à atteindre l'objectif

Ce cycle perception → raisonnement → action → observation est au cœur du fonctionnement de Claude Code.

## 2. Comment fonctionne un agent de code

Quand on soumet une tâche à Claude Code — par exemple "Analyse ce fichier CSV et produis un graphique de la consommation par bâtiment" — l'agent ne génère pas simplement du code. Il traverse un cycle de 5 phases.

### 2.1 Le modèle d'exécution en 5 phases

```
┌─────────────────────────────────────────────────────────────────┐
│                   CYCLE D'EXÉCUTION CLAUDE CODE                 │
│                                                                  │
│   DEMANDE UTILISATEUR                                            │
│   "Analyse ce CSV et fais un graphique"                          │
│              │                                                   │
│              ▼                                                   │
│   ┌──────────────────┐                                           │
│   │  1. PLANIFICATION │  → Décompose la tâche en étapes         │
│   │                  │    Identifie les outils nécessaires       │
│   └────────┬─────────┘    Évalue les risques                    │
│            │                                                     │
│            ▼                                                     │
│   ┌──────────────────┐                                           │
│   │   2. LECTURE     │  → Explore la structure du projet        │
│   │                  │    Lit les fichiers pertinents           │
│   └────────┬─────────┘    Comprend le contexte existant         │
│            │                                                     │
│            ▼                                                     │
│   ┌──────────────────┐                                           │
│   │   3. ÉCRITURE    │  → Génère le code adapté au contexte    │
│   │                  │    Crée ou modifie les fichiers          │
│   └────────┬─────────┘    Respecte les conventions du projet    │
│            │                                                     │
│            ▼                                                     │
│   ┌──────────────────┐                                           │
│   │  4. EXÉCUTION    │  → Lance le script Python                │
│   │                  │    Observe la sortie (résultats/erreurs) │
│   └────────┬─────────┘    Capture les messages d'erreur         │
│            │                                                     │
│            ▼                                                     │
│   ┌──────────────────┐                                           │
│   │  5. SYNTHÈSE     │  → Explique ce qui a été fait           │
│   │                  │    Présente les résultats clés           │
│   └────────┬─────────┘    Propose des améliorations             │
│            │                                                     │
│            ▼                                                     │
│   RÉSULTAT : fichiers créés + explication + suggestions         │
│                                                                  │
│   ↺ Si erreur détectée en phase 4 → retour phase 3 (itération) │
└─────────────────────────────────────────────────────────────────┘
```

### 2.2 Exemple concret du cycle sur une tâche réelle

**Demande** : "Calcule le taux de défauts moyen par ligne de production pour janvier 2024"

| Phase | Ce que fait Claude Code | Ce qu'on observe dans le terminal |
|-------|-------------------------|-----------------------------------|
| **1. Planification** | Identifie qu'il faut lire un CSV, filtrer par date, grouper par ligne, calculer un ratio | Affiche son raisonnement en texte |
| **2. Lecture** | Lit `production_industrielle.csv`, inspecte les colonnes, vérifie les types | `Reading file: production_industrielle.csv` |
| **3. Écriture** | Génère un script pandas avec les bonnes opérations | Crée `analyse_defauts.py` |
| **4. Exécution** | Lance `python analyse_defauts.py`, lit la sortie | Affiche les résultats ou une erreur |
| **5. Synthèse** | Explique les résultats, signale les anomalies détectées | Résumé en langage naturel |

Si en phase 4 une erreur de type de données est levée (ex : la colonne `date` est lue comme du texte), l'agent retourne en phase 3, corrige le parsing des dates, et ré-exécute. Tout cela sans intervention.

### 2.3 La gestion des erreurs — une capacité distinctive

Un point souvent sous-estimé : Claude Code ne produit pas un code parfait du premier coup, exactement comme un développeur humain. Ce qui le distingue, c'est sa capacité à **lire les messages d'erreur et à les corriger en autonomie**.

Exemple typique :

```
Erreur Python : KeyError: 'batiment'
→ L'agent lit l'erreur
→ Inspecte les colonnes réelles du CSV
→ Découvre que la colonne s'appelle 'Bâtiment' (avec accent)
→ Corrige le code
→ Ré-exécute
```

Ce cycle de correction itérative représente une grande partie de la valeur de l'outil pour des non-développeurs.

## 3. Les outils de l'agent

Claude Code dispose d'un ensemble d'outils spécialisés. Comprendre leur rôle permet de formuler des demandes plus précises et de mieux interpréter ce que fait l'agent.

### 3.1 Tableau des outils principaux

| Outil | Rôle | Quand il est utilisé | Exemple |
|-------|------|----------------------|---------|
| **Read** | Lire le contenu d'un fichier | Explorer un CSV, lire du code existant, lire une configuration | Lit `consommation_energie.csv` pour comprendre sa structure |
| **Write** | Créer un nouveau fichier ou réécrire intégralement | Créer un script d'analyse, générer un rapport HTML | Crée `analyse_energie.py` avec le code complet |
| **Edit** | Modifier une partie précise d'un fichier existant | Corriger un bug, ajuster un paramètre, ajouter une fonction | Modifie uniquement le titre d'un graphique dans un script existant |
| **Bash** | Exécuter une commande dans le terminal | Lancer Python, installer une bibliothèque, lister des fichiers | `python analyse_energie.py` ou `pip install plotly` |
| **Glob** | Lister des fichiers selon un pattern | Trouver tous les CSV d'un dossier, lister tous les scripts Python | `data/*.csv` → retourne la liste de tous les CSV |
| **Grep** | Rechercher du texte dans des fichiers | Trouver où une variable est utilisée, chercher une colonne dans du code | Cherche `kwh` dans tous les scripts pour comprendre les usages |

> **Note** : ces outils s'exécutent dans le répertoire de travail courant (le dossier depuis lequel Claude Code a été lancé). Il est donc important de lancer l'outil depuis la racine du projet.

### 3.2 Combinaison d'outils sur une tâche typique

Pour une demande comme "Crée un script qui analyse la production de Ligne_1 et génère un graphique", l'agent va combiner plusieurs outils en séquence :

```
Glob("data/*.csv")           → identifie les fichiers disponibles
Read("production.csv")       → comprend la structure (colonnes, types)
Write("analyse_ligne1.py")   → crée le script Python
Bash("python analyse.py")    → exécute le script
Read(sortie d'erreur)        → si erreur, lit le message
Edit("analyse_ligne1.py")    → corrige le code
Bash("python analyse.py")    → ré-exécute
```

Cette orchestration automatique des outils est ce qui distingue un agent d'un simple générateur de code.

## 4. Quand utiliser Claude Code

Claude Code n'est pas une réponse universelle à tous les besoins. Il excelle dans des contextes précis. Voici 5 cas d'usage concrets dans un environnement industriel.

### Cas d'usage 1 — Analyse exploratoire de données industrielles

**Contexte** : Un export CSV de 50 000 lignes arrive chaque semaine depuis la GTC (Gestion Technique Centralisée). Il contient des mesures de consommation énergétique par bâtiment, zone et type d'énergie. L'analyste doit produire un résumé chaque lundi matin.

**Prompt type** :
```
Analyse le fichier data/consommation_energie.csv. 
Calcule la consommation totale par bâtiment et par type d'énergie pour 2024.
Identifie les 3 postes de consommation les plus élevés.
Génère un graphique en barres et exporte les résultats en CSV.
```

**Ce que fait Claude Code** : explore le fichier, calcule les agrégats, génère le code pandas + plotly, produit les fichiers de sortie.

**Temps estimé sans outil** : 2 à 4 heures (formules Excel, copier-coller, mise en forme)
**Temps avec Claude Code** : 5 à 15 minutes

### Cas d'usage 2 — Détection d'anomalies sur données capteurs

**Contexte** : Une usine dispose de 12 capteurs de température répartis dans les zones de production. Les données sont exportées chaque nuit. L'ingénieur de maintenance souhaite être alerté quand un capteur sort de sa plage nominale.

**Prompt type** :
```
Analyse capteurs_temperature.csv. 
Pour chaque capteur, calcule la moyenne et l'écart-type sur les 30 derniers jours.
Identifie les mesures qui dépassent moyenne ± 2 écarts-types.
Génère un rapport HTML avec les graphiques des capteurs anormaux.
```

**Ce que fait Claude Code** : implémente la logique de détection statistique, génère les visualisations, produit un rapport exploitable directement.

**Valeur ajoutée** : sans compétences Python, cette tâche nécessiterait un développeur ou des jours de travail sous Excel.

### Cas d'usage 3 — Automatisation d'un rapport récurrent

**Contexte** : Chaque fin de mois, le responsable de production doit envoyer un rapport PDF avec les KPIs de production : taux de défauts par ligne, productivité par équipe, évolution mensuelle.

**Prompt type** :
```
Crée un script Python qui lit production_industrielle.csv,
calcule les KPIs du mois écoulé, génère les graphiques
et exporte un rapport PDF. 
Le script doit fonctionner avec python rapport_mensuel.py
sans aucune autre intervention.
```

**Ce que fait Claude Code** : construit un script réutilisable, gère les cas limites (données manquantes, mois sans production), documente le code.

**Valeur durable** : le script tourne ensuite de façon autonome chaque mois, sans refaire appel à l'agent.

### Cas d'usage 4 — Nettoyage et standardisation de données

**Contexte** : Les données de production proviennent de 3 systèmes différents avec des formats incompatibles — colonnes nommées différemment, dates dans des formats variés, valeurs manquantes, doublons.

**Prompt type** :
```
J'ai 3 CSV dans data/brut/ : système A, B et C.
Ils contiennent les mêmes données mais avec des formats différents.
Crée un script qui les unifie dans un format commun,
supprime les doublons, normalise les dates en ISO 8601,
et signale les valeurs manquantes dans un log.
```

**Ce que fait Claude Code** : analyse chaque fichier, identifie les mappings de colonnes, implémente la logique de fusion, génère un rapport de qualité des données.

### Cas d'usage 5 — Construction d'un dashboard opérationnel

**Contexte** : Un chef d'atelier souhaite un tableau de bord affiché sur un écran en production, se rafraîchissant toutes les heures avec les données du jour.

**Prompt type** :
```
Crée un dashboard Streamlit avec :
- KPIs du jour en haut (production totale, taux de défauts, temps d'arrêt)
- Graphique d'évolution horaire de la production
- Tableau des alertes (lignes au-dessus du seuil de défauts)
Les données viennent de production_industrielle.csv qui est mis à jour chaque heure.
```

**Ce que fait Claude Code** : construit l'application Streamlit complète, configure le rafraîchissement automatique, structure le code pour faciliter les modifications futures.

**Résultat** : une application web fonctionnelle en 30 minutes, sans écrire une ligne de code manuellement.

## 5. Ce que Claude Code ne fait pas

Comprendre les limites d'un outil est aussi important que comprendre ses capacités. Voici les frontières importantes.

### 5.1 Limites techniques

**Il ne se connecte pas à Internet en autonomie**
Claude Code ne peut pas aller chercher des données sur le web, appeler des APIs externes, ou envoyer des emails par lui-même. Il travaille uniquement avec les fichiers et outils disponibles localement (sauf si on lui fournit explicitement les outils pour le faire via des scripts).

**Il ne mémorise pas les sessions précédentes**
Chaque nouvelle session repart de zéro. Si on a travaillé sur un projet la veille, il faut lui re-contextualiser la situation (en lançant Claude Code depuis le bon dossier et en expliquant brièvement l'objectif). Il explore le projet pour comprendre l'existant, mais ne se souvient pas des conversations passées.

**Il n'a pas accès aux systèmes de l'entreprise**
Sans configuration spécifique, il ne peut pas lire les données de la GTC en temps réel, accéder à l'ERP, ou se connecter à une base de données SQL. Il travaille avec des fichiers exportés manuellement.

### 5.2 Limites de jugement

**Il ne vérifie pas la pertinence métier des résultats**
Si les données sont incorrectes (mauvaise unité, valeur aberrante non signalée, période mal filtrée), Claude Code produira des résultats techniquement corrects mais métier faux. La validation des résultats reste de la responsabilité de l'utilisateur.

**Il peut se tromper**
Claude Code n'est pas infaillible. Il peut mal interpréter une demande ambiguë, générer du code avec des bugs subtils, ou produire une analyse statistiquement incorrecte. Tout résultat critique doit être vérifié.

**Il ne remplace pas l'expertise métier**
Il peut calculer un taux de défauts, mais il ne sait pas si ce taux est acceptable pour ce type de ligne de production. L'interprétation des résultats demande toujours un expert du domaine.

### 5.3 Ce qu'il vaut mieux faire autrement

| Situation | Pourquoi Claude Code n'est pas optimal | Alternative |
|-----------|----------------------------------------|-------------|
| Mise en page d'un rapport Word/PowerPoint | Il peut générer du contenu mais pas mettre en forme visuellement | LibreOffice, Word, PowerPoint |
| Analyse ad hoc sur 10 lignes de données | Le temps de setup dépasse le bénéfice | Excel |
| Déploiement en production | Nécessite des processus qualifiés, tests, validation | DevOps, SI |
| Décisions stratégiques | L'IA fournit de l'information, pas du jugement | Humains qualifiés |

## 6. Le modèle de permissions

Claude Code a accès au système de fichiers et peut exécuter des commandes. Il est important de comprendre comment ce système fonctionne pour travailler en toute confiance.

### 6.1 Périmètre d'accès aux fichiers

Claude Code accède uniquement au dossier depuis lequel il est lancé et à ses sous-dossiers. Si on lance Claude Code depuis `/home/utilisateur/projets/analyse-energie/`, il pourra lire et écrire dans ce dossier, mais pas accéder à `/home/utilisateur/documents/` ni à d'autres dossiers du système.

```
Dossier de lancement : /projets/analyse-energie/
    ✓ Accessible : /projets/analyse-energie/data/consommation.csv
    ✓ Accessible : /projets/analyse-energie/scripts/
    ✗ Inaccessible : /projets/autre-projet/
    ✗ Inaccessible : /documents/confidentiel/
```

**Bonne pratique** : créer un dossier de projet dédié pour chaque analyse, et lancer Claude Code depuis ce dossier.

### 6.2 Permissions d'exécution

Claude Code peut exécuter des commandes système via l'outil Bash. Il demande une confirmation avant d'exécuter des commandes potentiellement dangereuses (suppression de fichiers, installation de logiciels).

Le mode par défaut (`--dangerously-skip-permissions`) peut être activé pour des environnements de confiance (poste personnel, environnement isolé), ce qui supprime les demandes de confirmation. **Ce mode n'est pas recommandé sur des systèmes partagés ou en production.**

| Type d'action | Comportement par défaut |
|---------------|------------------------|
| Lire un fichier | Automatique |
| Créer un nouveau fichier | Automatique |
| Modifier un fichier existant | Automatique |
| Supprimer un fichier | Demande confirmation |
| Installer un package Python | Demande confirmation |
| Exécuter un script Python | Automatique |
| Commandes réseau | Demande confirmation |

### 6.3 Traçabilité des modifications

Tous les fichiers créés ou modifiés par Claude Code sont de simples fichiers texte sur le disque. Il est donc possible :

- De voir exactement ce qui a été créé avec un explorateur de fichiers
- D'utiliser Git pour versionner les modifications et revenir en arrière
- De lire et comprendre le code généré ligne par ligne
- De modifier manuellement ce que l'agent a produit

Il n'y a pas de "boîte noire" : tout ce que fait Claude Code est visible, lisible, et modifiable.

### 6.4 Données et confidentialité

Les données des fichiers lus par Claude Code sont envoyées à l'API Anthropic pour traitement. Pour des données sensibles :

- Vérifier la politique de confidentialité d'Anthropic et les conditions de l'offre utilisée (API, Claude for Work)
- Pour des données RGPD ou des données industrielles confidentielles, envisager une offre entreprise ou un déploiement on-premise
- Il est possible d'anonymiser les données avant de les soumettre à l'agent pour l'exploration

> **Règle de bon sens** : ne pas soumettre de données personnelles identifiables ou de données classifiées sans vérification préalable des conditions contractuelles.
