## Étape 4 : Python et uv

Python est le langage d'analyse de données utilisé dans cette formation. `uv` est un gestionnaire de projets Python moderne — il remplace `pip` et `venv` avec une approche plus simple et plus rapide.

### Installation de Python 3.12+

```python
# ÉTAPE 4a — Vérifier si Python est déjà installé
import subprocess, sys

result = subprocess.run(["python3", "--version"], capture_output=True, text=True)
if result.returncode == 0:
    print("Python trouvé :", result.stdout.strip())
    # Vérifier la version minimale
    version = sys.version_info
    if version.major == 3 and version.minor >= 12:
        print("Version compatible (3.12+)")
    else:
        print(f"Version actuelle : {version.major}.{version.minor} — une mise à jour vers 3.12+ est recommandée")
else:
    print("Python non trouvé — installation requise")
```

### Pourquoi uv plutôt que pip ?

| Aspect | pip + venv (méthode classique) | uv (méthode recommandée) |
|--------|--------------------------------|--------------------------|
| **Vitesse d'installation** | 30-60s pour pandas | 2-5s pour pandas |
| **Gestion de l'environnement virtuel** | Commandes séparées (`python -m venv .venv`, puis `source .venv/bin/activate`) | Intégré : `uv init` crée tout |
| **Fichier de dépendances** | `requirements.txt` à maintenir manuellement | `pyproject.toml` géré automatiquement |
| **Conflits de versions** | Possibles, difficiles à diagnostiquer | Résolution automatique |
| **Compatibilité** | Universelle | Tous systèmes depuis 2024 |

## Étape 5 : Vérification complète

Ce script vérifie que tous les composants sont correctement installés. Exécuter la cellule suivante pour un diagnostic complet.

```python
# Script de vérification complète de l'environnement
# Exécuter cette cellule pour diagnostiquer l'installation

import subprocess
import sys
import importlib

VERT  = "\033[92m"
ROUGE = "\033[91m"
JAUNE = "\033[93m"
RESET = "\033[0m"

def verifier_commande(nom, commande):
    """Vérifie qu'une commande est accessible dans le terminal."""
    try:
        result = subprocess.run(commande, capture_output=True, text=True)
        version = result.stdout.strip()
        print(f"{VERT}[OK]{RESET} {nom} : {version}")
        return True
    except FileNotFoundError:
        print(f"{ROUGE}[MANQUANT]{RESET} {nom} : non installé")
        return False

def verifier_module_python(nom_module):
    """Vérifie qu'un module Python est importable."""
    try:
        module = importlib.import_module(nom_module)
        version = getattr(module, "__version__", "version inconnue")
        print(f"{VERT}[OK]{RESET} Python module '{nom_module}' : {version}")
        return True
    except ImportError:
        print(f"{ROUGE}[MANQUANT]{RESET} Python module '{nom_module}' : non installé")
        print(f"        -> Installer avec : uv add {nom_module}")
        return False

print("=" * 55)
print("  VÉRIFICATION DE L'ENVIRONNEMENT DATA MASTERY")
print("=" * 55)
print()

print("--- Commandes système ---")
ok_node   = verifier_commande("Node.js",     ["node",   "--version"])
ok_npm    = verifier_commande("npm",         ["npm",    "--version"])
ok_claude = verifier_commande("Claude Code", ["claude", "--version"])
ok_uv     = verifier_commande("uv",          ["uv",     "--version"])

print()
print("--- Python ---")
print(f"{VERT}[OK]{RESET} Python : {sys.version.split()[0]}")
version = sys.version_info
if version.major == 3 and version.minor >= 12:
    print(f"{VERT}[OK]{RESET} Version Python >= 3.12")
else:
    print(f"{JAUNE}[AVERTISSEMENT]{RESET} Python {version.major}.{version.minor} — 3.12+ recommandé")

print()
print("--- Bibliothèques Python ---")
ok_pandas  = verifier_module_python("pandas")
ok_plotly  = verifier_module_python("plotly")
ok_jupyter = verifier_module_python("jupyter")

print()
print("=" * 55)
tous_ok = ok_node and ok_npm and ok_uv and ok_pandas and ok_plotly
if tous_ok:
    print(f"{VERT}Environnement complet — prêt pour la formation.{RESET}")
else:
    print(f"{JAUNE}Des composants sont manquants — suivre les étapes ci-dessus.{RESET}")
print("=" * 55)
```

## Étape 6 : Premier test — une session complète

Ce premier test valide que tout fonctionne ensemble : Claude Code peut lire des fichiers, générer du code Python, et l'exécuter.

```python
# ÉTAPE 6a — Créer la structure du projet de test
# Ces commandes créent un projet minimal avec un CSV d'exemple

# Dans le terminal :
# mkdir ~/projets/test-claude
# cd ~/projets/test-claude
# uv init .
# uv add pandas plotly
# mkdir data

# On peut aussi faire ça depuis Python pour vérifier :
import pandas as pd

# Créer un CSV de test minimal
donnees_test = pd.DataFrame({
    "mois":       ["Jan", "Fév", "Mar", "Avr", "Mai", "Juin"],
    "production": [1200, 1350, 1100, 1450, 1380, 1290],
    "defauts":    [24,   18,   31,   12,   22,   19]
})

# Sauvegarder dans un CSV temporaire pour le test
chemin_test = "/tmp/test_production.csv"
donnees_test.to_csv(chemin_test, index=False)
print(f"CSV de test créé : {chemin_test}")
print(donnees_test.to_string(index=False))
```

```python
# ÉTAPE 6b — Exemple du prompt à donner à Claude Code
# Dans le terminal, depuis le dossier du projet :
# claude

# Puis taper ce prompt :
prompt_exemple = """
Lis le fichier data/test_production.csv.
Calcule le taux de défauts pour chaque mois (defauts / production * 100).
Crée un graphique en barres avec plotly qui affiche ce taux.
Sauvegarde le graphique dans output/taux_defauts.html.
"""

print("Prompt à donner à Claude Code dans le terminal :")
print("-" * 50)
print(prompt_exemple)
print("-" * 50)
```

```python
# ÉTAPE 6c — Le code que Claude Code générerait pour cette tâche
# Voici un exemple représentatif du script produit :
import pandas as pd
import plotly.express as px

# Chargement du CSV
df = pd.read_csv("/tmp/test_production.csv")

# Calcul du taux de défauts
df["taux_defauts_pct"] = (df["defauts"] / df["production"] * 100).round(2)

# Création du graphique
fig = px.bar(
    df,
    x="mois",
    y="taux_defauts_pct",
    title="Taux de défauts par mois",
    labels={"taux_defauts_pct": "Taux de défauts (%)", "mois": "Mois"}
)

# Affichage
fig.show()
print("Taux de défauts calculés :")
print(df[["mois", "taux_defauts_pct"]].to_string(index=False))
```

## Résolution de problèmes

Voici les problèmes les plus fréquents lors de l'installation, avec leurs solutions.

| Problème | Message d'erreur typique | Solution |
|----------|--------------------------|----------|
| **Node.js introuvable** | `command not found: node` | Redémarrer le terminal après l'installation ; vérifier que le PATH inclut le dossier Node.js |
| **npm introuvable** | `command not found: npm` | npm est inclus avec Node.js — réinstaller Node.js |
| **Claude Code introuvable** | `command not found: claude` | Redémarrer le terminal ; si persistant, vérifier `npm list -g @anthropic-ai/claude-code` |
| **Erreur de permissions npm** | `EACCES permission denied` | Sur Mac/Linux : `sudo npm install -g @anthropic-ai/claude-code` ou reconfigurer npm |
| **Clé API invalide** | `AuthenticationError` | Vérifier la clé sur console.anthropic.com ; la clé doit commencer par `sk-ant-` |
| **uv introuvable** | `command not found: uv` | Redémarrer le terminal ; ajouter `~/.cargo/bin` au PATH si demandé par l'installeur |
| **Module pandas manquant** | `ModuleNotFoundError: No module named 'pandas'` | Depuis le dossier du projet : `uv add pandas` |
| **Erreur SSL / certificat** | `SSL: CERTIFICATE_VERIFY_FAILED` | Sur macOS : lancer `/Applications/Python 3.12/Install Certificates.command` |
| **Python non trouvé** | `command not found: python3` | Installer Python 3.12 depuis python.org ; sur Windows, cocher "Add to PATH" |
| **Jupyter ne s'ouvre pas** | Port 8888 déjà utilisé | `jupyter lab --port 8889` pour utiliser un autre port |
